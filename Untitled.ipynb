```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#переменные
#каждую ячейку запускаем через shift+enter

#средняя 
m = 180
#sigma
s = 20

#минимальное значение зп
ms_min = m - s*3
#максимальное значение зп
ms_max = m + s*3

#заголовок
my_title = "Middle PHP Developer"
#название оси
ox = 'тыс. руб. в месяц net'
```

```python
x = np.linspace(ms_min,ms_max,100000)
y = stats.norm.pdf(x,m,s)

ax = plt.axes([0,0,1,1], frameon=False)
ax.get_xaxis().set_visible(True)
ax.get_yaxis().set_visible(False)

plt.xlabel (ox, color='gray')
plt.plot(x,y,c="grey", fillstyle = 'full')
plt.title(my_title)


plt.fill_between(x, y, where=(x>ms_max-s), color='r', alpha=0.6)
plt.text (ms_max-3*s/4, 0.0001, '2,3%', color='white', style= 'oblique')

plt.fill_between(x, y, where=(x>ms_max-2*s), color='r', alpha=0.4)
plt.text (ms_max-7*s/4, 0.0025, '13,6%', color='white', style= 'oblique')

plt.fill_between(x, y, where=(x>ms_max-3*s), color='r', alpha=0.2)
plt.text (ms_max-11*s/4, 0.01, '34,1%', color='red', style= 'oblique')

plt.fill_between(x, y, where=(x<ms_min+3*s), color='r', alpha=0.2)
plt.text (ms_min+9*s/4, 0.01, '34,1%', color='red', style= 'oblique')

plt.fill_between(x, y, where=(x<ms_min+2*s), color='r', alpha=0.4)
plt.text (ms_min+ 5*s/4, 0.0025, '13,6%', color='white', style= 'oblique')

plt.fill_between(x, y, where=(x<ms_min+s), color='r', alpha=0.6)
plt.text (ms_min+s/2, 0.0001, '2,3%', color='white', style= 'oblique')

plt.show()
```
